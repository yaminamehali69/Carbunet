```python
for col_name, col_type in columns:
    if col_type.startswith("character") or col_type == "text":
        query = f"SELECT COUNT(*) FROM consommation_energie WHERE {col_name} IS NOT NULL AND {col_name} <> '';"
    else:
        query = f"SELECT COUNT(*) FROM consommation_energie WHERE {col_name} IS NOT NULL;"
    
    cur.execute(query)
    count = cur.fetchone()[0]
    
    # Afficher même si la colonne est vide
    if count == 0:
        print(f"Colonne vide : {col_name}")
    else:
        print(f"Colonne {col_name} a {count} valeurs non nulles")
```

```python
# Récupérer les colonnes de la table
cur.execute("""
SELECT column_name, data_type
FROM information_schema.columns
WHERE table_name = 'consommation_energie';
""")
columns = cur.fetchall()

# Générer et exécuter une requête pour compter les valeurs égales à zéro
for col_name, col_type in columns:
    if col_type.startswith("character") or col_type == "text":
        # Pour les colonnes de type texte, on ignore les comparaisons avec 0
        query = f"SELECT COUNT(*) FROM consommation_energie WHERE {col_name} = '';"
    else:
        # Pour les autres types de données, on cherche les valeurs égales à zéro
        query = f"SELECT COUNT(*) FROM consommation_energie WHERE {col_name} = 0;"
    
    cur.execute(query)
    count = cur.fetchone()[0]
    
    if count > 0:
        print(f"Colonne {col_name} a {count} valeurs égales à zéro.")
```

```python
import pandas as pd

# Lire le fichier avec l'encodage le plus probable
df = pd.read_csv(r"C:\Users\mehal\Desktop\Yamina\vsc\backup_consommation_energie.csv", encoding="utf-8", low_memory=False)

# Corriger les chaînes avec un double décodage
def reencode(val):
    if isinstance(val, str):
        try:
            return val.encode('latin1').decode('utf-8')
        except:
            return val
    return val

# Appliquer la correction
df = df.map(reencode)

# Tester le résultat
print(df["filiere"].unique())
print(df["libelle_commune"].head())
```
